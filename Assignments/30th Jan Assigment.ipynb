```python
percent = int(input("Enter percentage"))

if percent>90:
    Grade = "A"
elif (percent > 80) & (percent <=90):
    Grade = "B"
elif (percent >=60) & (percent <= 80):
    Grade = "C"
elif percent < 60:
    Grade = "D"
print(f"Grade for {percent} percent is : ",Grade)
```

```python
cost = int(input("Enter bike cost"))

if cost>100000:
    Tax = "15%"
elif (cost>50000) & (cost<=100000):
    Tax = "10%"
elif cost <50000:
    Tax = "5%"
print(f"tax to be paid for {cost} cost is : ",Tax)
```

```python
city = input("Enter City").capitalize()

if city=="Delhi":
    Monument = "Red Fort"
elif city == "Agra":
    Monument = "Taj Mahal"
elif city == "Jaipur":
    Monument="Jal Mahal"

print(f"City {city} has monument {Monument}")
```

```python
num = int(input("Enter a num"))
times = 0
while num >10:
    num = num - 3
    times = times + 1
print(f"Times the number {num} is divisible by 3 before it is less than or equal to 10: {times}")
```
