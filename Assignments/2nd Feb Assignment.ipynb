1. Tuples are immutable unlike lists
2. count() - counts the occurence of the particular num or obj <br> index() - returns the position of the num or obj
3. Set() datatype doesnt allow duplicate items

```python
tup = (1,2,3,3,4)
list = [1, 1, 1, 2, 1, 3, 1, 4, 2, 1, 2, 2, 2, 3, 2, 4, 3, 1, 3, 2, 3, 3, 3, 4, 4, 1, 4, 2, 4, 3, 4, 4]
print("counts function : ",tup.count(2))
print("index function : ", tup.index(3))
print("Set dataype removes duplicates : ", set(list))
```

4. update() updates the original set and returns None, whereas union() leaves the original set in place and returns a new set
5. items are ordered in a ditionary. eg. of ditionary {'a':2, 'b':3}
6. example of nested dictionary : {'a':{'apple':2, 'banana':4}}
7. dict1.setdefault('topic', ['Python', 'Machine Learning', 'Deep Learning'])

```python
dat = {1,2,3,3,4}
dat.update({5})
print("after update() ",dat)

dat2 = dat.union({6})
print("original unchanged after union :",dat)
print("after union :",dat2)

# Nested Dictionary
dic = {'a':{'apple':2, 'banana':4}}
print("nested dictionary:",dic['a'])

# key named topics
dict1 = {'language' : 'Python', 'course': 'Data Science Masters'}
dict1.setdefault('topic', ['Python', 'Machine Learning', 'Deep Learning'])
print(dict1)
```
